## PyCity Schools Analysis

```python
#import dependencies
import pandas as pd
```

```python
#load csv's to use
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
```

```python
#have pandas read the csv's
schools = pd.read_csv(school_data)
students = pd.read_csv(student_data)
students.head()
```

```python
#combine data into one set
all_data = pd.merge(students, schools, how="left", on=["school_name", "school_name"])
all_data.head()
```

```python
df_all = pd.DataFrame(all_data)
df_all.head()
```

## District Summary

```python
df_schools = pd.DataFrame(schools)
schools_count = df_schools["school_name"].value_counts()
total_schools_count = schools_count.count()
total_schools_count
```

```python
# Calculate the total number of students
st_count = df_all["Student ID"].value_counts()
student_count = st_count.count()
student_count
```

```python
# Calculate the total budget
total_budget = df_schools["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = round((df_all["math_score"].mean()), 2)
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = round((df_all["reading_score"].mean()), 2)
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = all_data[(all_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / (student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = all_data[(all_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / (student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = all_data[
    (all_data["math_score"] >= 70) & (all_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = {"Total Schools": [total_schools_count], 
                    "Total Students": [student_count], 
                    "Total Budget": [total_budget], 
                    "Average Math Score": [average_math_score],
                   "Average Reading Score": [average_reading_score], 
                    "% Passing Math": [passing_math_percentage], 
                    "% Passing Reading": [passing_reading_percentage],
                   "% Overall Passing": [overall_passing_rate]}

district_summary_df = pd.DataFrame(district_summary)

# Formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:,.2f}%".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:,.2f}%".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:,.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:,.2f}%".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:,.2f}%".format)

#Display the DataFrame
district_summary_df
```

## School Summary

```python
# Use the code provided to select the school type
school_types = schools.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count
per_school_counts = schools.set_index(["school_name"])["size"]
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = all_data.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
per_school_capita
```

```python
# Calculate the average math test scores
per_school_math = all_data.groupby(["school_name"]).mean()["math_score"]
per_school_math
```

```python
# Calculate the average reading test scores
per_school_reading = all_data.groupby(["school_name"]).mean()["reading_score"]
per_school_reading
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = all_data[all_data["math_score"] >= 70]
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = all_data[all_data["reading_score"] >= 70]
school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
overall_passing = all_data[(all_data["reading_score"] >= 70) & (all_data["math_score"] >= 70)]
overall_passing
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
overall_passing_rate = overall_passing.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = {"School Type": school_types,
                      "Total Students": per_school_counts, 
                      "Budget": per_school_budget, 
                    "Per Student Budget": per_school_capita, 
                      "Average Math Score": per_school_math,
                   "Average Reading Score": per_school_reading, 
                    "Math Passing Rates": per_school_passing_math, 
                      "Reading Passing Rates": per_school_passing_reading,
                   "Overall Passing Rate": overall_passing_rate}

per_school_summary_df = pd.DataFrame(per_school_summary)

#Format
per_school_summary_df["Total Students"] = per_school_summary_df["Total Students"].map("{:,}".format)
per_school_summary_df["Budget"] = per_school_summary_df["Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)
per_school_summary_df["Average Math Score"] = per_school_summary_df["Average Math Score"].map("{:.2f}%".format)
per_school_summary_df["Average Reading Score"] = per_school_summary_df["Average Reading Score"].map("{:.2f}%".format)
per_school_summary_df["Math Passing Rates"] = per_school_summary_df["Math Passing Rates"].map("{:,.2f}%".format)
per_school_summary_df["Reading Passing Rates"] = per_school_summary_df["Reading Passing Rates"].map("{:,.2f}%".format)
per_school_summary_df["Overall Passing Rate"] = per_school_summary_df["Overall Passing Rate"].map("{:,.2f}%".format)

# Display the DataFrame
per_school_summary_df
```
